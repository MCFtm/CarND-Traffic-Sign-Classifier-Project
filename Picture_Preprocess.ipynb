```python
from PIL import Image, ImageOps
import os
import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline
```

```python
input_folder='./traffic-signs-data-collected-input/'
output_folder='./traffic-signs-data-collected-output/'
```

```python
imagelist=os.listdir(input_folder)
size = (32, 32)
```

```python
for i in imagelist:
    img = Image.open(input_folder+i)
    img.thumbnail(size, Image.ANTIALIAS)
    background = Image.new('RGBA', size, (0, 0, 0, 255))
    background.paste(img, (int((size[0] - img.size[0]) / 2), int((size[1] - img.size[1]) / 2)))
    background.save(output_folder+i[: -4]+'-output.png')
```

```python
imagelist=os.listdir(output_folder)
for i,img in enumerate(imagelist):
    # plt.subplot(19, 5, ((i//3)*3*5)+(i%3*2)+1) #row-id:((i//3)*2*5) ;column: (i%3*2)+1
    # plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.subplot(7,10,(i%3*3)+1)
    image = Image.open(output_folder+img)
    plt.imshow(image)
    plt.title(img[: -4])
    plt.axis('on')
    if (i+1)%3==0:
        plt.show()
```
